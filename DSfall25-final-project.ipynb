# Section 2: Exploritory Data Analysis

The big question that I want to ask here is: is Art Garfunkel's taste in books common or unusual by the standards of the Goodreads users? I will find this out by comparing the median number of Goodreads ratings on Art Garfunkel's favorites to the median number of Goodreads ratings of the overall Goodreads dataframe. I plan to use the median as my measure of center because there are some books that are absurdly popular and some that no one has read yet and these outliers would majorly skew the results of the mean. The higher the median is, the more popular of books Garfunkel has read.

```python
# rename columns to match the goodreads dataset
df_ArtG = df_Art.rename(columns = {"Date Read": "date_read", "Author": "authors", "Books": "title", "Year Published": "published_year", "Pages": "num_pages", "Favorite": "favorite"})

#change published_year datatype to be more easily comparable
df_ArtG["published_year"].astype(int)
df_ArtG
```

```python
# remove books that were giving formatting errors and are not included in the Art dataframe
df_filteredGoodreads = df_Goodreads[(df_Goodreads.authors != "Elizabeth  George") & (df_Goodreads.language_code == "eng")]

# reformat the publication date column to be more 
df_filteredGoodreads["publication_date"] = pd.to_datetime(df_filteredGoodreads["publication_date"]) 
# create column that includes only published year
df_filteredGoodreads["published_year"] = pd.DatetimeIndex(df_filteredGoodreads["publication_date"]).year

# drop rows that have the same title so that there is only one instance of each book
df_Books = df_filteredGoodreads.drop_duplicates(subset = 'title', keep = 'first')
#get rid of books with less than 100 ratings because those are not reliable ratings
df_Books = df_Books[df_Books["ratings_count"] >= 100]
df_Books
```

```python
# get rid of columns that don't have anything to do with popularity
df_books = df_Books.drop(columns = ["isbn", "isbn13", "  num_pages", "authors", "published_year", "publication_date", "bookID", "language_code"]) 
df_both = pd.merge(df_books, df_ArtG, how="inner", on = ["title"])
df_both
```

```python
# mean, median, standard deviation of the Goodreads dataframe
books_median = df_Books["ratings_count"].median()
```

```python
# mean, median, standard deviation of Garfunkel's favorites
df_both_favorites = df_both[df_both["favorite"] == 1]
both_median = df_both_favorites["ratings_count"].median()
```

```python
# display overall stats
print("median # of Goodreads reviews:", books_median)
```

```python
# display Garfunkel's favorites' stats
print("median # of Goodreads reviews of Garfunkel's favorites:", both_median)
```

If a book is more popular, it will probably have more ratings on Goodreads, since more people have read it. The median number of reviews of the overall Goodreads dataset is 2359 which gives us the baseline that your typical book will have a number of ratings around 2359. This is important because, for Garfunkel's favorites, the median number of reviews is far higher at 6995 reviews. This tells us that the books that Garfunkel reads are usually more popular than your typical book.

# Section 5: Data Science

In section 4, I pondered if I would be able to create a pseudo-recommendation algorithm using k-means, but noted that I did not have enough information. Using an additional dataset, data,  that includes genres, I believe that a rudimentary recommendation algorithm is now possible.  

My question is: based on his listed favorites, what other books would Art Garfunkel enjoy? I will answer this by using the kNeighbors Classifier introduced in Microproject 11 to first train the data on Garfunkel's favorite books and then apply the trained model to other books from the supplementary data set. Additionally, to incorporate strings into my machine learning, I will turn the different genres/authors into columns in which boolean values indicate which book(s) have that genre/author.

```python
import pandas as pd
```

```python
# load book data and drop unnecessary columns
df_Art = pd.read_csv("Art Garfunkel Library.csv")
df_Art = df_Art.rename(columns = {"Date Read": "date_read", "Author": "authors", "Books": "title", "Year Published": "published_year", "Pages": "num_pages", "Favorite": "favorite"})
df_Art = df_Art.drop(columns = ["date_read", "published_year", "num_pages"])

df_data = pd.read_csv("data.csv")
df_data = df_data.drop(columns = ["isbn10", "isbn13", "thumbnail", "description", "subtitle"])
df_data = df_data.drop_duplicates(subset = 'title', keep = 'first')


df_ArtData = pd.merge(df_data, df_Art, how="inner", on = ["title", "authors"])
df_ArtData
```

```python
# create dataframe that associates categories with numbers
categories = pd.DataFrame(df_data["categories"].drop_duplicates())
categoryID  = pd.get_dummies(categories[['categories']], dtype = int)
categoryID["categories"] = categories
```

```python
# create dataframe that associates authors with numbers
authors = pd.DataFrame(df_data["authors"].drop_duplicates())
authorID  = pd.get_dummies(authors[['authors']], dtype = int)
authorID["authors"] = authors
```

```python
modeldata = pd.merge(categoryID, df_ArtData, how="inner", on = "categories")
modeldata = pd.merge(authorID, modeldata, how="inner", on = "authors")
modeldata["average_rating"].astype(int)
modeldata["num_pages"].astype(int)
modeldata.info()
```

```python
trainingdata = modeldata.tail(156)
trainingdata = trainingdata.reset_index(drop = True)
testdata = modeldata.head(30)
testdata = testdata.reset_index(drop = True)
```

```python
ind = trainingdata.drop(columns = ["authors", "categories", "title", "published_year", "average_rating", "num_pages", "ratings_count", "favorite"])
ind = list(ind.columns)
ind.append("average_rating")
ind.append("num_pages")
ind.append("ratings_count")
dep = "favorite"

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

model.fit(trainingdata[ind], trainingdata[dep])
```

```python
testdata["reccomended"] = model.predict(testdata[ind])
testdata[((testdata["favorite"] == 0) & (testdata["reccomended"] == 1)) | ((testdata["favorite"] == 1) & (testdata["reccomended"] == 0))]
```

```python
df_reccomended = df_data
df_reccomended = pd.merge(categoryID, df_data, how="inner", on = "categories")
df_reccomended = pd.merge(authorID, df_reccomended, how="inner", on = "authors")
df_reccomended = df_reccomended.dropna()
ind2 = df_reccomended.drop(columns = ["authors", "categories", "title", "published_year", "average_rating", "num_pages", "ratings_count"])
ind2 = list(ind2.columns)
ind2.append("average_rating")
ind2.append("num_pages")
ind2.append("ratings_count")
df_reccomended["reccomended"] = model.predict(df_reccomended[ind2])
```

```python
passed = df_reccomended[df_reccomended["reccomended"] == 1]
passed[["title", "authors"]]
```

# Section 6: A Different Data Visualization

```python
def noOutliers(df, col):                  # eliminate outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = (Q3-Q1)
    lValues = Q1 - 1.5 * IQR
    hValues = Q3 + 1.5 * IQR
    df_recplot = df[(df[col] > lValues) & (df[col] < hValues)] 
    return df_recplot

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rating = noOutliers(df_reccomended, "average_rating")

sns.boxplot(data = rating, y = "average_rating",  x = "reccomended").set(title = "Comparative average rating of books seperated by if they would be recommended")

plt.ylabel("Average rating")
plt.xlabel("Recommended to Garfunkel where 0 is not recommended and 1 is recommended")
```

```python
ratehist = rating["average_rating"].hist(bins = 50)
ratehist.set_title("Frequency of average ratings")
ratehist.set_xlabel("Average Rating")
ratehist.set_ylabel("Frequency")
```

```python
pages = noOutliers(df_reccomended, "num_pages")
sns.boxplot(data = pages, y = "num_pages",  x = "reccomended").set(title = "Comparative number of pages of books seperated by if they would be recommended")

plt.ylabel("Number of Pages")
plt.xlabel("Recommended to Garfunkel where 0 is not recommended and 1 is recommended")
```

```python
pagehist = pages["num_pages"].hist(bins = 50)
pagehist.set_title("Frequency of different page numbers")
pagehist.set_xlabel("Number of Pages")
pagehist.set_ylabel("Frequency")
```

```python
popularity = noOutliers(df_reccomended, "ratings_count")
sns.boxplot(data = popularity, y = "ratings_count",  x = "reccomended").set(title = "Comparative number of ratings of books seperated by if they would be recommended")

plt.ylabel("Number of Ratings")
plt.xlabel("Recommended to Garfunkel where 0 is not recommended and 1 is recommended")
```

```python
pagehist = popularity["ratings_count"].hist(bins = 50)
pagehist.set_title("Frequency of different numbers of ratings")
pagehist.set_xlabel("Number of Ratings")
pagehist.set_ylabel("Frequency")
```

I used these graphs to demonstrate how the machine learning classifier used three variables to determine whether a book should be recommended to Art Garfunkel. This is not a measure of accuracy, mearly a way to see what kind of books were put into each category. In the first box plot, we can see that average ratings do not differ much between the two groups, but that is to be expected since there is not too much variation in how the books are rated overall which we can see in the histogram of the average ratings. This was likely not the most determining factor for the classifier.   

In the next box plot, we learn that while the median number of pages for recommended and non-recommended books are similar while the box portion of the box plots overlap in a relatively small area. This makes sense when we look at the histogram for the number of pages since there is a more even distribution which would lend itself well to one category being more significantly higher than the other. The classifier seems to have placed books with more pages into Garfunkel's recommendation pile more frequently than books with comparatively fewer pages.    

In the final box plot, we can observe that while the books with more reviews were recommended more, the non-recommended books had far more outliers in the positive direction. There are so many books with zero reviews (which we can see very clearly in the histogram where the smallest bin has over 2000 books in it) just because of quantity, new edidions getting less attention, etc. that it makes it difficult to have a box plot that is both readable and has similar data to the classifier since it pulls everything down. This makes it even more significant that the classifier tended to recommend books that had more reviews than the median of not recommended books. This was likely a large contributing factor to the classification of each book.
